```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
points = np.random.uniform(0, 5, (100, 2))
```

```python
a = 0.6666666666
b = 1.0
c = -3.0
```

```python
colors = []
for p in points:
  colors.append('blue' if a * p[0] + b * p[1] + c >= 0 else 'red')
```

```python
plt.scatter(points[:, 0], points[:, 1], c=colors)
```

```python
W = np.random.uniform(-1.0, 1.0, 3)
```

```python
W
```

```python
def predict(W, X):
  return 1.0 if W[0] * X[0] + W[1] * X[1] + W[2] * X[2] >= 0 else -1.0
```

```python
def prepend_x0(X):
  return np.array([1.0, *X])
```

```python
points_to_predict = []
predicted_colors = []
for x in np.arange(0, 5, 5.0/ 100.0):
  for y in np.arange(0, 5, 5.0/100.0):
    point = [x, y]
    points_to_predict.append(point)
    X = prepend_x0(point)
    predicted_colors.append('lightblue' if predict(W, X) == 1.0 else 'pink')
points_to_predict = np.array(points_to_predict)
plt.scatter(points_to_predict[:, 0], points_to_predict[:, 1], c=predicted_colors)

plt.scatter(points[:, 0], points[:, 1], c=colors)
```
